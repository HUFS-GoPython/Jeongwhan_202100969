#### list comprehension
- 포함형/지능형/축약형 리스트
- 하나의 리스트에서 그걸 포함해 다른 리스트를 만듦
- list + for loop
- [A for A' in C]
    - C에 있는 A'에서 A를 가져와라 => 리스트에 담아서 반환
    - A : 내가 원하는 출력 형태

    - C : 반복가능한 자료형 (리스트)
 
- 조건 추가 : [A for A' in C if ...]

### NLTK
- Frequency Distribution : 단어가 몇 번 출현했는지 (빈도수)
- for 텍스트 성질 이해 
    <= 단어 개수
    <= 빈도수
실현 : `FreqDist(text)`

```python
fdist = FreqDist(text1) #cased
```

```python
fdist
```

```python
fdist.most_common(50) #가장 빈번한 상위 50개
```

```python
fdist['that'], text1.count('that') #둘이 같음
```

```python
fdist.freq('that')
```

```python
fdist.plot(50, cumulative=True)
```

```python
#가장 빈번하지 않은 = 1번만 나오는 단어들
 # 알파벳만 리스트에 담아보기, 중복없이
len(fdist.hapaxes())
```

```python
[word for word in fdist.hapaxes() if word.isalpha()]
```

```python
# 대소문자 구분 없이 freq dist 만드려면? the와 The.. 내꺼랑 비교해보기 
fdist2 = FreqDist([word.lower() for word in text1])
fdist2['the']
```

```python
# text 2에서 길이가 3자 이하이고 빈도수가 5 이상

#FreqDist(text2)>=5
[word for word in text2 if len(word)<=3 and text2.count(word)>=5]
```
